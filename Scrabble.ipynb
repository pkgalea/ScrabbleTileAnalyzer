```python
def get_player_name_and_id(link):
    return link.text, link['href'].split("p=")[1]

def get_non_linked_player_names(player_names):
    players_split = player_names.split(" vs. ")
    if (len(players_split)==1):
        p1_name = ""
        p2_name = ""
    else:
        p1_name = players_split[0]
        p2_name = players_split[1].rstrip()
    return p1_name, p2_name
```

```python
   
#conn = pg2.connect(user='postgres',  dbname='scrabble', host='localhost', port='5432', password='myPassword')

for p in player_collection.find():
    print(p['p_id'])
#    print(p['content'])
    soup = BeautifulSoup(p['content'], 'html.parser')
    content_block = soup.find("table", {"class": "playercontentblock"})
    p_id = p['p_id']
    name = content_block.find("div", {"class": "playername"}).text.strip()
#    print(p_id)
    name = name.split("|")[0]
    stats = [d.find("div", {"class": "stat"}) for d in content_block.findAll("div", {"class":"playerheaderinfo"})]
    rating = stats[0].text
    rating = -1 if rating=="---" else stats[0].text
    rank = stats[1].text.strip()
    rank = -1 if rank=="(unranked)" else stats[1].text[:-2]
    win_loss_tie = stats[2].text.split("-")
    wins = int(win_loss_tie[0])
    losses = int(win_loss_tie[1])
    ties = int(win_loss_tie[2].split("(")[0])
#    print(wins, losses, ties)
    avg_score = stats[3].text.strip()
    avg_score_top, avg_score_bottom = -1, -1 if avg_score == "---" else avg_score.split('-')
    print(p_id, name, rating, rank, wins, losses, ties, avg_score_top, avg_score_bottom)
```
